```python
def post_aim_api(url,link="", data = {'title': 'dummy title', 'id': 1, 'message': 'hello world!'}, headers = {'Content-Type': 'application/json; chearset=utf-8'}):
    res = requests.post(url+link, data=json.dumps(data), headers=headers) 
    print(str(res.status_code) + " | " + res.text)
    return res
```
